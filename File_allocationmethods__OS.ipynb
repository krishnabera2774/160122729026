#### Exp 4 - File Allocation using Contiguous, Linked and Indexed Methods

### Contiguous allocation

```python
class ContiguousAllocation:
    def __init__(self, total_blocks):
        self.memory = [None] * total_blocks

    def allocate(self, file_name, start_block, num_blocks):
        for i in range(num_blocks):
            self.memory[start_block + i] = file_name

    def display_memory(self):
        print(self.memory)


total_blocks = 100
contig_alloc = ContiguousAllocation(total_blocks)
contig_alloc.allocate("file1", start_block=10, num_blocks=5)
contig_alloc.display_memory()
```

### Linked list Allocation

```python
class LinkedAllocation:
    def __init__(self):
        self.memory = {}

    def allocate(self, file_name, block_list):
        self.memory[file_name] = block_list

    def display_memory(self):
        for file_name, block_list in self.memory.items():
            print(f"{file_name}: {block_list}")


linked_alloc = LinkedAllocation()
linked_alloc.allocate("file2", [2, 5, 8])
linked_alloc.display_memory()
```

### Indexed Allocation

```python
class IndexedAllocation:
    def __init__(self, total_blocks):
        self.memory = [None] * total_blocks
        self.index_blocks = {}

    def allocate(self, file_name, block_list):
        index_block = len(self.index_blocks)
        self.index_blocks[index_block] = block_list
        for block in block_list:
            self.memory[block] = (file_name, index_block)

    def display_memory(self):
        print("Memory:")
        print(self.memory)
        print("Index Blocks:")
        print(self.index_blocks)


total_blocks = 50
indexed_alloc = IndexedAllocation(total_blocks)
indexed_alloc.allocate("file3", [12, 15, 18])
indexed_alloc.display_memory()
```

#####Exp 06 - Implementation of producer-consumer, readers- writers and dining philosophers problem using semaphores

```python
## Producer Consumer problem

import threading
import time

BUFFER_SIZE = 5
buffer = []
mutex = threading.Semaphore(1)
empty = threading.Semaphore(BUFFER_SIZE)
full = threading.Semaphore(0)

def producer():
    while True:
        item = "item"
        empty.acquire()
        mutex.acquire()
        buffer.append(item)
        print(f"Produced: {item}")
        mutex.release()
        full.release()
        time.sleep(1)

def consumer():
    while True:
        full.acquire()
        mutex.acquire()
        item = buffer.pop(0)
        print(f"Consumed: {item}")
        mutex.release()
        empty.release()
        time.sleep(1)

# Create producer and consumer threads
producer_thread = threading.Thread(target=producer)
consumer_thread = threading.Thread(target=consumer)

producer_thread.start()
consumer_thread.start()
```

```python
## Writing reading

readers = 0
mutex = threading.Semaphore(1)
resource = threading.Semaphore(1)

def reader():
    global readers
    mutex.acquire()
    readers += 1
    if readers == 1:
        resource.acquire()
    mutex.release()


    print("Reading...")

    mutex.acquire()
    readers -= 1
    if readers == 0:
        resource.release()
    mutex.release()

def writer():
    resource.acquire()

    print("Writing...")

    resource.release()


for _ in range(3):
    threading.Thread(target=reader).start()

threading.Thread(target=writer).start()
```

```python
## Dining philosphers problem
import threading
NUM_PHILOSOPHERS = 5
chopsticks = [threading.Semaphore(1) for _ in range(NUM_PHILOSOPHERS)]

def philosopher(i):
    while True:

        print(f"Philosopher {i} is thinking")

        chopsticks[i].acquire()
        chopsticks[(i + 1) % NUM_PHILOSOPHERS].acquire()


        print(f"Philosopher {i} is eating")

        chopsticks[i].release()
        chopsticks[(i + 1) % NUM_PHILOSOPHERS].release()


for i in range(NUM_PHILOSOPHERS):
    threading.Thread(target=philosopher, args=(i,)).start()
```
